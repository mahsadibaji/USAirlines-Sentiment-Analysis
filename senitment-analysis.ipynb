# **Functions for data processing**

```python
def lex_norm():
    norm_dict = {}
    with open(r"C:\Users\Mahsa\Downloads\Sem 99-00\NLP\emnlp2012-lexnorm\emnlp_dict.txt") as f:
        for line in f:
            (key, val) = line.split()
            norm_dict[key] = val
    return norm_dict
for k,v in lex_norm().items():
    print(f"{k} : {v}")
```

```python
def remove_punctuation(words):
    """Remove punctuation from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = re.sub("[^a-zA-Z]", '', word)
        if new_word != '':
            new_words.append(new_word)
    return new_words
```

```python
def formalize(words):
    """
    informal language normalization a list of tokenized words
    """
#     print("informal language normalization a list of tokenized words")
    lexical_norm=lex_norm()
    formalized_words=[lexical_norm[word] if word in lexical_norm else word for word in words]
    
    return formalized_words
```

```python
def remove_stopwords(words):
    """
    Remove stopwords from list of tokenized words
    """
#     print("Remove stopwords from list of tokenized words")
    words=[w for w in words if w not in stopwords.words('english')]

    return words
```

```python
def remove_URL_and_mentions(text):
    """Remove URLs from a text"""
#     print("Remove URLs from a text")
    return re.sub(r"(?:\@|RT @|https?\://)\S+", "", text)
```

```python
def replace_contractions(text):
    """Replace contractions in string of text"""
#     print("Replace contractions in string of text")
    return contractions.fix(text)
```

```python
def remove_short_words(words):
#     print("remove short words")
    words=[w for w in words if len(w)>=3]
    
    return words
```

```python
def stemmer(words):
    ps = PorterStemmer() 
    words = [ps.stem(w) for w in words]
    return words
```

# **Data Loading**

```python
train_data=pd.read_csv(r"C:\Users\Mahsa\Downloads\Sem 99-00\NLP\airline-train.csv")
validation_data=pd.read_csv(r"C:\Users\Mahsa\Downloads\Sem 99-00\NLP\airline-dev.csv")
test_data=pd.read_csv(r"C:\Users\Mahsa\Downloads\Sem 99-00\NLP\airline-test.csv")

train_data['target'] = train_data['airline_sentiment'].map({'positive':1,'neutral':0,'negative':-1})
validation_data['target'] = validation_data['airline_sentiment'].map({'positive':1,'neutral':0,'negative':-1})
test_data['target']=test_data['airline_sentiment'].map({'positive':1,'neutral':0,'negative':-1})
```

# **Data PreProcessing:**
**in this step each data set is cleaned using preprocess function.**

**1.remove urls,mentions,RT keywords.**

**2.Replace contractions(you've --> you have).**

**3. Remove puncuations**

**4.formal language normalization is done using emnlp lexicon.**

**5.stop words are removed with nltk stop word list.**

**6.stemming using nltk stemmer and in the end short words (length< 3) are removed**

```python
def preprocess(d):
    """preprocess a comment"""
        
    d=remove_URL_and_mentions(d)

    d=replace_contractions(d)

    d=d.lower()
    
    d=word_tokenize(d)
       
    d=remove_punctuation(d)
        
    d=formalize(d)

    d=remove_stopwords(d)
    
    d=stemmer(d) 
    
    d=remove_short_words(d)
    
    d=" ".join(w for w in d)
    
    return d
```

```python
train_data['text']=train_data['text'].apply(preprocess)  
```

```python
validation_data['text']=validation_data['text'].apply(preprocess)
```

```python
test_data['text']=test_data['text'].apply(preprocess)
```

```python
x_train=train_data['text']
y_train=train_data['target']

x_val=validation_data['text']
y_val=validation_data['target']

x_test=test_data['text']
y_test=test_data['target']
```

# **Feautre Selection:**
**using TfidfVectorizer and chi2 from sklearn,features are sorted by their chi square measure meaning.**

```python
tvec = TfidfVectorizer(use_idf=True)
x_train_tfidf = tvec.fit_transform(x_train)
x_val_tfidf=tvec.transform(x_val)
x_test_tfidf=tvec.transform(x_test)
```

```python
chi2score = chi2(x_train_tfidf, y_train)[0]
```

**the following graph shows the most effective features using chi2 measurement (first 20)**

```python
plt.figure(figsize=(15,10))
wscores = zip(tvec.get_feature_names(), chi2score)
wchi2 = sorted(wscores, key=lambda x:x[1])
topchi2 = zip(*wchi2[-20:])
lst_topchi2=list(topchi2)
x = range(len(lst_topchi2[1]))
labels = lst_topchi2[0]
plt.barh(x,lst_topchi2[1], align='center', alpha=0.2)
plt.plot(lst_topchi2[1], x, '-o', markersize=5, alpha=0.8)
plt.yticks(x, labels)
plt.xlabel('$\chi^2$')
```

# **Train Classifier version 1**
for the first version of classifier, we selected 400 features using selectKbest function.
then with knn classifier and train_data, our model is trained.

```python
from sklearn.neighbors import KNeighborsClassifier
ch2 = SelectKBest(chi2, k=400)
x_train_chi2_selected = ch2.fit_transform(x_train_tfidf, y_train)
x_val_chi2_selected=ch2.transform(x_val_tfidf)
knn = KNeighborsClassifier(n_neighbors=5, weights='distance').fit(x_train_chi2_selected, y_train)
```

```python
accuracy = knn.score(x_val_chi2_selected, y_val)
print(accuracy)
```

**Results for classifier v1**

```python
from sklearn.metrics import classification_report,confusion_matrix
knn_pred=knn.predict(x_val_chi2_selected)
print(classification_report(y_val,knn_pred))
```

**Confusion matrix of classifer version1**

```python
import seaborn as sn

cm = confusion_matrix(y_val, knn_pred)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)

plt.show()
```

# Hyperparameters tuning

```python
k_values=[1000, 2000,3000,4000,5000,6000]
n_neighbors_values=[2,4,6,8,10]
results=[]
for k in k_values:
    for n in n_neighbors_values:
        ch2_tuning = SelectKBest(chi2, k=k)
        x_train_chi2_selected = ch2_tuning.fit_transform(x_train_tfidf, y_train)
        x_val_chi2_selected=ch2_tuning.transform(x_val_tfidf)
        knn = KNeighborsClassifier(n_neighbors=n, weights='distance').fit(x_train_chi2_selected, y_train)
        accuracy = knn.score(x_val_chi2_selected, y_val)
        results.append((k,n,accuracy))
        print (f"#selected features: {k}\n#neighbors: {n}\naccuracy: {accuracy}")
        print("-"*10)
```

# **Train final classifier **
using parameters that yeild the best result among others

```python
from operator import itemgetter
max_accracy=max(results,key=itemgetter(2))

ch2 = SelectKBest(chi2, k=max_accracy[0])
x_train_chi2_selected = ch2.fit_transform(x_train_tfidf, y_train)
x_val_chi2_selected=ch2.transform(x_val_tfidf)
knn = KNeighborsClassifier(n_neighbors=max_accracy[1], weights='distance').fit(x_train_chi2_selected, y_train)
```

# Results on test data:

```python
x_test_chi2_selected=ch2.transform(x_test_tfidf)
prediction=knn.predict(x_test_chi2_selected)
print(classification_report(y_test,prediction))
```

```python
cm = confusion_matrix(y_test, prediction)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)

plt.show()
```
